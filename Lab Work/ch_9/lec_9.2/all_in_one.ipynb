## Task 1: Stack Plot
Create a stack plot to visualize the sales data of three products (A, B, and C) over 12 months. Use different colors for each product and label the axes and the plot.

```python
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)
product_A = np.random.randint(10, 50, size=12)
product_B = np.random.randint(20, 60, size=12)
product_C = np.random.randint(15, 55, size=12)

plt.stackplot(months, product_A, product_B, product_C, labels=['Product A','Product B','Product C'])
plt.legend(loc='upper left')
plt.title('Sales Data of Products A, B, C Over 12 Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

## Task 2: Stacked Plot for Revenue
Given data of monthly revenue from four departments (HR, Marketing, Sales, IT), create a stacked plot to show the revenue distribution for each department over 6 months.

```python
months = np.arange(1, 7)
HR = np.random.randint(10, 30, size=6)
Marketing = np.random.randint(15, 35, size=6)
Sales = np.random.randint(20, 40, size=6)
IT = np.random.randint(25, 45, size=6)

plt.stackplot(months, HR, Marketing, Sales, IT, labels=['HR','Marketing','Sales','IT'])
plt.legend(loc='upper left')
plt.title('Monthly Revenue Distribution by Department')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

## Task 3: Subplot
Create a subplot with two graphs: one for a bar plot and another for a line plot. Label the axes, include titles for each plot, and adjust the layout so the plots do not overlap.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))

x = np.arange(1, 6)
y = np.random.randint(10, 50, size=5)

ax1.bar(x, y, color='skyblue')
ax1.set_title('Bar Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Values')

ax2.plot(x, y, marker='o', color='red')
ax2.set_title('Line Plot')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Values')

plt.tight_layout()
plt.show()
```

## Task 4: Line Plot with Legend
Create a line plot showing the temperature variation over a year for three different cities. Include a legend to label the cities and place the legend at the upper-left corner.

```python
months = np.arange(1, 13)
city1 = np.random.randint(15, 35, size=12)
city2 = np.random.randint(10, 30, size=12)
city3 = np.random.randint(5, 25, size=12)

plt.plot(months, city1, label='City 1')
plt.plot(months, city2, label='City 2')
plt.plot(months, city3, label='City 3')
plt.legend(loc='upper left')
plt.title('Temperature Variation Over a Year')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.show()
```

## Task 5: Save Bar Chart
Save a bar chart showing the distribution of ages of students in a class. Ensure the chart is saved as a PNG file on your system.

```python
ages = np.random.randint(10, 20, size=30)
plt.hist(ages, bins=5, color='orange', edgecolor='black')
plt.title('Distribution of Ages of Students')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('student_ages.png')
plt.show()
```

## Task 6: Save Pie Chart
Create a pie chart showing the market share of 5 companies. Save the chart as a PNG file and display it using Matplotlib.

```python
companies = ["A", "B", "C", "D", "E"]
market_share = np.random.randint(10, 40, size=5)

plt.pie(market_share, labels=companies, autopct="%1.1f%%", startangle=90, colors=["gold", "tomato", "skyblue", "#76ffb4", "plum"])
plt.title("Market Share of Companies")
plt.savefig("market_share.png")
plt.show()
```
