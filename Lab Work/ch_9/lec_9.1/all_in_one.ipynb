# Matplotlib Charts

## Task 1: Create a Simple Bar Chart
Create a bar chart showing sales data for five products.

```python
import matplotlib.pyplot as plt

products = ["A", "B", "C", "D", "E"]
sales = [150, 200, 300, 250, 180]

plt.bar(products, sales)
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales Data for Five Products")
plt.show()
```

## Task 2: Create a Line Chart
Plot monthly temperature changes in a city for 12 months.

```python
months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr",
    "May",
    "Jun",
    "Jul",
    "Aug",
    "Sep",
    "Oct",
    "Nov",
    "Dec",
]
temperatures = [30, 32, 35, 40, 45, 43, 38, 36, 34, 33, 31, 29]

plt.plot(months, temperatures, marker="o")
plt.xlabel("Months")
plt.ylabel("Temperature (C)")
plt.title("Monthly Temperature Changes")
plt.show()
```

## Task 3: Create a Scatter Plot
Relationship between hours studied and exam scores for 20 students.

```python
import numpy as np

hours_studied = np.random.randint(1, 10, 20)
exam_scores = hours_studied * 10 + np.random.randint(-5, 5, 20)

plt.scatter(hours_studied, exam_scores)
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Hours Studied vs Exam Scores")
plt.show()
```

## Task 4: Create a Pie Chart
Percentage distribution of different fruits sold in a store.

```python
fruits = ["Apples", "Bananas", "Cherries", "Dates"]
quantities = [50, 30, 10, 10]

plt.pie(quantities, labels=fruits, autopct="%1.1f%%", startangle=90)
plt.title("Fruit Sales Distribution")
plt.show()
```

## Task 5: Create a Histogram
Display the frequency distribution of ages of 100 people.

```python
ages = np.random.randint(10, 70, 100)

plt.hist(ages, bins=10, edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

## Task 6: Combined Chart
Plot both bar and line chart together showing revenue and profit for 6 months.

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
revenue = [200, 250, 300, 280, 350, 400]
profit = [50, 70, 90, 80, 100, 120]

plt.bar(months, revenue, label="Revenue")
plt.plot(months, profit, color="red", marker="o", label="Profit")
plt.xlabel("Months")
plt.ylabel("Amount")
plt.title("Revenue and Profit")
plt.legend()
plt.show()
```

## Task 7: Customized Charts
Create a bar chart with customized colors, labels, and grid lines.

```python
products = ["A", "B", "C", "D", "E"]
sales = [150, 200, 300, 250, 180]

plt.bar(products, sales, color=["red", "blue", "green", "orange", "purple"])
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Customized Bar Chart")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

## Task 8: Annotate Charts
Create a line chart and annotate the highest and lowest points.

```python
values = [10, 15, 20, 25, 18, 30, 22]
plt.plot(values, marker="o")
plt.title("Line Chart with Annotations")

max_val = max(values)
min_val = min(values)
plt.annotate(
    "Highest",
    xy=(values.index(max_val), max_val),
    xytext=(values.index(max_val), max_val + 2),
    arrowprops=dict(facecolor="green", shrink=0.05),
)
plt.annotate(
    "Lowest",
    xy=(values.index(min_val), min_val),
    xytext=(values.index(min_val), min_val - 4),
    arrowprops=dict(facecolor="red", shrink=0.05),
)
plt.show()
```
